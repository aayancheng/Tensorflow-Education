```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# using numpy to generate random data and form a simple linear function for target
x_data = np.random.rand(100).astype(np.float32)
y_data = x_data * 0.2 + 0.3
```

```python
plt.scatter(x_data, y_data)
plt.show()
```

```python
# define tensorflow variable and assign an initial guess of variables and then apply linear model to make predictions (inference)
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b
```

```python
#  define loss and optimizer for training
loss = tf.reduce_mean(tf.square(y - y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)
```

```python
# Before starting, initialize the variables
init = tf.global_variables_initializer()
```

```python
# Launch the graph
sess = tf.Session()
sess.run(init)
```

```python
# Fit the line
for step in range(501):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(W), sess.run(b))
```
